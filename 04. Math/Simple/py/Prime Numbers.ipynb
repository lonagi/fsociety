```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
_Ns, _Xs, _Bs, _As, _Xmins=[],[],[],[],[]
```

```python
for n in range(301,911+1,10):
    if(n!=301):
        break
    
    xlist=[]
    Blist=[]
    Alist=[]
    b1_x=[]
    xts=[]
    
    for x in range(1,int((n-1)/2)+1):
        B=((n-(n//x)*x)**2)//x
        A=int(((n-x*(int(n/x)+1))**(2))/x)
        xlist.append(x)
        #Blist.append(B)
        #Alist.append(A)
        
        Blist.append(B*x+2)
        Alist.append(A*x+2)
        
        xt=(x*B+2)**(1/2)
        if(xt-round(xt)==0):
            xts.append(xt)
        else:
            pass
            xts.append(0)
        
        
        
        #print(f'n={n}: x={x} -> B={B}')
        #print(f'n={n}: x={x} -> A={A}')
        if(B==1):
            #print(f'n={n}: x={x} -> B=1')
            b1_x.append(x)
        
    Bmax=max(Blist)
    Xmax=xlist[Blist.index(Bmax)]
    
    #print(f"n={n}: X={Xmax} -> B={Bmax}")
    #print(f'n={n}: x_min={min(b1_x)} -> B=1')
    
    if(len(xts)==0):
        pass
        #print(f'n={n}: x_min={min(b1_x)} -> нет п.квадратов')
    
    _Xmins.append(min(b1_x))
    _Ns.append(n)
    _Xs.append(Xmax)
    _Bs.append(Bmax)
print("Done")
```
